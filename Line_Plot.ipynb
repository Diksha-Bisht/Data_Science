# pyplot is used to create figures and change their characteristics.

# %matplotlib inline function allows for plots to be visible when using Jupyter Notebook.

```python
# Installing Matplotlib
""" pip install matplotlib """
# Importing the libraries
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd                       #for creating data frames
import numpy as np
```

creating data and converting it to dataframe using pandas


```python
data= {'Days':['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
       'Week1':[23,54,23,12,33,44,34],
       'Week2':[43,23,12,43,23,54,23],
       'Week3':[23,45,23,43,22,34,23]}
df=pd.DataFrame(data)                   #Converting Dictionary to dataframe
df
```

ploting the graph

```python
plt.plot(df['Days'],df['Week1'])
```

```python
x=df['Days']
y=df['Week1']
plt.plot(x,y,color='green', marker='o', linestyle= '--')
plt.title('Days VS Temp plot of Week 1')
plt.xlabel('Days')
plt.ylabel('Temperaature')
```

```python
x = np.linspace(0, 10, 100)  # Example x-axis values (e.g., time)
y = np.sin(x)               # Example y-axis values (e.g., some function of x)
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)   #label gives the name to the graph which is shown in the legend
plt.legend()
```

#*More on* **markers**


```python
plt.plot(x,y,'bo--')
```

```python
plt.plot(x,y, 'rd--')
```

use of legend and

'c' for color to manually set the color for each line and

ls  for line style


```python
w1= plt.plot(df['Days'],df['Week1'],label='Week-1',c='red',ls='--')
w2=plt.plot(df['Days'],df['Week2'],label='Week-2', c='green', ls='-')
w3= plt.plot(df['Days'], df['Week3'],label='Week-3', c='purple', ls='--')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Days VS Temperature plot of All weeks")
plt.legend(loc='lower right')
plt.grid(True)                #Adds grid to graph
```

#Subplots


```python
pl=plt.subplot()
pl.plot(x,y)
pl.spines['top'].set_visible(False)
pl.spines['right'].set_visible(False)
```
