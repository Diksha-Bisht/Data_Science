#Bar Plot

```python
import matplotlib.pyplot as plt
```

plt.bar(x, height, width, bottom, align)

```python
x=[2,5,3,4,6,7,3]
y=[2,4,1,5,6,2,3]

plt.bar(x,y)
plt.title('Simple Bar Plot')
plt.figure(figsize=(10,5))
```

```python
d={'Maths':23,'Science':12,'English':26,'Hindi':29}
courses=list(d.keys())
marks=list(d.values())

plt.bar(courses,marks)
plt.title('Bar plot with dict to list pickup')
```

#working with dataframes

```python
import pandas as pd

# Storing Data to csv file
Date=['10-03-21','10-04-21','10-05-21','10-06-21','10-07-21']
Open=[774.5, 776.030029, 779.309998, 779, 779.659973]
High=[776.065002, 778.710022, 782.070007, 780.4799, 779.65]
Low=[769.5, 772.89, 775.65, 776.859985,775.75]
Close=[772.55999, 776.42999, 776.46997, 776.8599, 775.080017]

dict={'Date':Date,'Open':Open,'High':High,'Low':Low,'Close':Close}

#Creating DataFrame
df=pd.DataFrame(dict)

#Converting to .csv
df.to_csv('xyz.csv')
```

```python
#Extracting data from data frame
d=df['Date']
o=df['Open']

plt.bar(d[:3],o[:3])      #Working with top 3 of data
plt.ylim(770, 780)
plt.title('Working with dataframe')
```

Creating horizontal bar plot

```python
plt.barh(d,o)
plt.xlim(770,780)
plt.title('Horizontal Bar Plot')
```

#subplots and formatting

```python
# adding watermark
fig, ax = plt.subplots(figsize =(5, 5))     #creating subplots of size 5*5 and storing it into fig whith axes in ax
ax.barh(d,o)
plt.xlim(770,780)
plt.title('Watermarked Bar Plot(Subplot)')
fig.text(0.9, 0.15, 'db', fontsize = 12, color ='red', ha ='right', va ='bottom', alpha = 0.7)           #adding watermark
```

removing spines

```python
import matplotlib.pyplot as plt

# Sample data
d = [1, 2, 3, 4, 5]
o = [10, 20, 15, 25, 30]

# Create a figure and axis
fig, ax = plt.subplots()

# Create the bar plot
ax.bar(d, o)

# Hide the spines (top, bottom, left, right)
for s in ['top', 'right']:
    ax.spines[s].set_visible(False)

#ax.xaxis.set_ticks_position('none')              These remove the ticks made on each label and axis
#ax.yaxis.set_ticks_position('none')

# Show the plot
plt.show()
```

#Multiple Bar Plot

```python
import numpy as np
x=[2,4,2,5,9,8]
y=[3,6,12,5,3,6]
z=[3,7,9,5,10,5]

bw=0.25

br1=np.arange(len(x))
br2=[a + bw for a in br1]
br3=[a + bw for a in br2]

plt.bar(br1,x,color='r',width=bw,label='IT')
plt.bar(br2,y,color='b',width=bw,label='ECE')
plt.bar(br3,z,color='c',width=bw,label='CSE')

plt.legend()
plt.title('Multiple Bar Plot')
plt.xlabel('Subjects')
plt.ylabel('Number of students')
```

#Stacked Bar Plot

```python
g=[20,30,22,31,28,32]
b=[10,20,23,29,14,21]
g2=[30,24,15,19,25,33]
b2=[20,35,33,29,10,19]
width=0.5

ind=np.arange(6)
plt.bar(ind,b,width,yerr=b2)      #yerr=> y error i.e, Error that might exist=> black lin in which actual value may fall
plt.bar(ind,g,width,bottom=b,yerr=g2)
plt.legend(['boys','girls'])
plt.title('Stacked Bar Graph')
plt.xlabel('Number')
plt.ylabel('Value')
```

changing value on x axis

```python
plt.bar(ind,b,width,yerr=b2)
plt.bar(ind,g,width,bottom=b,yerr=g2)
plt.legend(['boys','girls'])
plt.title('Stacked Bar Graph-X Value Changed')
plt.xlabel('Number')
plt.ylabel('Value')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5', 'T6'))   #Changing value on x-axis
plt.show()
```
