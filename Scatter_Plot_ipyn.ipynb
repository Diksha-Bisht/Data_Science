#Scatter Plot


```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rand
```

basic plot for single dataset

```python
x=[2,4,2,3,1,4,5]
y=[1,3,2,5,6,4,8]
plt.scatter(x,y)      #basic scatter plot

plt.title("Simple scatter plot")
```

Comparision for 2 datasets


```python
x1=[2,4,1,4,3,5]
y1=[3,5,7,2,6,7]
plt.scatter(x1,y1,c='r',marker='d',edgecolors='y')

x2=[3,4,2,4,2,3]
y2=[5,4,2,6,3,4]
plt.scatter(x2,y2,c='b',marker='+')

plt.title("Scatter for 2 Dataset with legend(column vice)")
plt.legend(['set1','set2'],ncol=2,loc='upper center')
```

```python
x1=[2,4,1,4,3,5]
y1=[3,5,7,2,6,7]
plt.scatter(x1,y1,c='r',marker='d',edgecolors='y')

x2=[3,4,2,4,2,3]
y2=[5,4,2,6,3,4]
plt.scatter(x2,y2,c='b',marker='+')

plt.title("Scatter for 2 Dataset with legend(outside)")
plt.legend(['set1','set2'],loc='upper center', bbox_to_anchor=(1.2,1)) # bbox_to_anchor ==> (x_right_distance, y_distance)
```

multicolor dots in a single plot for single dataset

```python
x=[1,3,2,4,2,4]
y=[4,2,5,2,6,2]
color=np.array(["red","green","blue","yellow","pink","black"])    # number of argument of input and color must be same
plt.scatter(x,y, c=color)

plt.title("Using multiple color marker")
```

using cmap (colormap) for giving multiple color in scatter marker

```python
color=np.array([0,2,4,29,40,100])
plt.scatter(x,y,c=color,cmap='viridis')

plt.title("Using cmap to give multiple color to marker")
```

```python
color=np.array([0,2,4,29,40,100])
plt.scatter(x,y,c=color,cmap='viridis')

plt.title("Using cmap (with colorbar) to give multiple color to marker")
plt.colorbar()          #showing viridis bar along with graph
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)     # Generate some sample data
y = np.sin(x)

cmap = plt.get_cmap('viridis')  # Create a colormap

colors = cmap(y)                # Create an array of colors based on the y-values

plt.figure(figsize=(8, 6))        # fixing map size
plt.scatter(x, y, c=colors, cmap=cmap)

plt.colorbar(label='Color by Value')    # color bar with label

plt.title('Colormaps with array based on y values')
```

Controlling marker size

```python
x=[2,4,3,5,2,5,3,4,2,5,7,3,7,8,9,5]
y=[2,4,2,5,7,2,6,5,4,3,4,4,6,4,3,2]
size=np.array([12,30,54,71,28,50,67,10,15,30,20,22,13,15,22,43])

plt.scatter(x,y,s=size)
plt.title("Variable size marker")
```

adjusting visiblility of markers

```python
x=[2,4,3,5,2,5,3,4,2,5,7,3,7,8,9,5]
y=[2,4,2,5,7,2,6,5,4,3,4,4,6,4,3,2]
size=np.array([12,30,54,71,28,50,67,10,15,30,20,22,13,15,22,43])

plt.scatter(x,y,s=size,alpha=0.5)
plt.title("Variable size with visibility= 50%")
```

```python
x=rand.randint(10, size=(100))
y=rand.randint(10,size=(100))
color=rand.randint(100,size=(100))
sizes=10*rand.randint(50,size=(100))

plt.scatter(x,y,s=sizes,c=color, alpha=0.3,cmap='nipy_spectral')
plt.colorbar()
plt.title("Diffrent size,color and opacity=30%")
```

#Scatter connected with line

```python
x=[1,3,4,2,5,6]
y=[2,4,5,2,5,3]
color=np.array([2,34,56,23,12,66])

plt.scatter(x,y,marker='d',c=color, cmap='nipy_spectral')
plt.plot(x,y,'r-.')
plt.colorbar()
plt.title("scater with line plot")
```

converting plot style to seaborn style

```python
x=rand.randint(10, size=(100))
y=rand.randint(10,size=(100))
color=rand.randint(100,size=(100))
sizes=10*rand.randint(50,size=(100))

plt.scatter(x,y,s=sizes,c=color, alpha=0.3,cmap='nipy_spectral')
plt.colorbar()
plt.title("Diffrent size,color and opacity=30%(Seaborn Style)")
plt.style.use("seaborn")
```
