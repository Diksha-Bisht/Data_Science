```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,20))
plt.subplots(2,2)   #Create 2*2 subplot group
plt.show()
```

```python
x=[1,2,3,4,5]
square=[1,4,9,16,25]
cube=[1,8,27,64,125]

square1=[1,3,8,15,24]
cube1=[1,7,26,63,124]

plt.subplot(2,2,1)    #1st part of 2*2 subplot group
plt.plot(x,square)

plt.subplot(2,2,2)    #2nd part of 2*2 subplot group
plt.bar(x,cube)

plt.subplot(2,2,3)    #3rd part of 2*2 subplot group
plt.scatter(x,square1)

plt.subplot(2,2,4)    #4th part of 2*2 subplot group
plt.plot(x,cube1)
plt.show()


#Savefig method
```
